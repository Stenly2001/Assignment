```python
import scipy.stats as stats
from statsmodels.stats.weightstats import ztest
```

```python
df = pd.read_csv("/content/Sales_add.csv")
df
```

```python
# 2 a.
alpha = 0.05
z_val,p_val = ztest(df['Sales_before_digital_add(in $)'],df['Sales_After_digital_add(in $)'])
if p_val<alpha:
  print("reject null hypothesis : increase in scale after stepping in digital market")
else:
  print("donot reject null hypothesis : no increase in scale after stepping in digital market")
```

```python
# 2 b.
import statsmodels.api as sm
from statsmodels.formula.api import ols
formula = "df['Sales_before_digital_add(in $)'] ~ C(Region)+C(Manager)+C(Region):C(Manager)"
model = ols(formula,data=df).fit()
anova_table = sm.stats.anova_lm(model,typ =2)
print(anova_table)
```

```python
alpha = 0.05
if any(anova_table["PR(>F)"]<alpha):
  print("there is signiificant dependency btw factors")
  reg =anova_table.loc["C(Region)","PR(>F)"]< alpha
  man = anova_table.loc["C(Manager)","PR(>F)"]< alpha
  if reg:
    print("there is no dependency between the features Region and Manager")
else:
       print("there is no dependency between the features Region and Manager")

```
