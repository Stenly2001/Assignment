```python
train_data['Gender'].fillna(method="ffill",inplace=True)
train_data["Married"].fillna(method="ffill",inplace=True)
train_data["Dependents"].fillna(method="ffill",inplace=True)
train_data['Self_Employed'].fillna(method="ffill",inplace=True)
train_data['LoanAmount'].fillna(method="bfill",inplace=True)
train_data['Loan_Amount_Term'].fillna(method="ffill",inplace=True)
train_data['Credit_History'].fillna(method="ffill",inplace=True)
```

```python
train_data.isnull().sum()
```

```python
test_data['Gender'].fillna(method="ffill",inplace=True)
test_data["Dependents"].fillna(method="ffill",inplace=True)
test_data['Self_Employed'].fillna(method="ffill",inplace=True)
test_data['LoanAmount'].fillna(method="bfill",inplace=True)
test_data['Loan_Amount_Term'].fillna(method="ffill",inplace=True)
test_data['Credit_History'].fillna(method="ffill",inplace=True)
```

```python
test_data.isnull().sum()
```

```python
train_data["Dependents"] = train_data["Dependents"].str.rstrip('+')
train_data["Dependents"] = train_data["Dependents"].astype('int64')
```

### Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
train_data["Gender"] = le.fit_transform(train_data["Gender"])
train_data["Married"] = le.fit_transform(train_data["Married"])
train_data["Education"] = le.fit_transform(train_data["Education"])
train_data["Self_Employed"] = le.fit_transform(train_data["Self_Employed"])
train_data["Property_Area"] = le.fit_transform(train_data["Property_Area"])
train_data["Loan_Status"] = le.fit_transform(train_data["Loan_Status"])
```

```python
train_data.head()
```

```python
train = train_data.drop('Loan_ID',axis =1)
```

```python
from sklearn import preprocessing
```

```python
std = preprocessing.StandardScaler()
```

```python
cols = ["ApplicantIncome","CoapplicantIncome","LoanAmount","Loan_Amount_Term"]
```

```python
for col in cols:
  train[col] = std.fit_transform(train[col].values.reshape(-1,1))
```

```python
train.head()
```

```python
test_data["Dependents"] = test_data["Dependents"].str.rstrip('+')
test_data["Dependents"] = test_data["Dependents"].astype('int64')
```

```python
test_data["Gender"] = le.fit_transform(test_data["Gender"])
test_data["Married"] = le.fit_transform(test_data["Married"])
test_data["Education"] = le.fit_transform(test_data["Education"])
test_data["Self_Employed"] = le.fit_transform(test_data["Self_Employed"])
test_data["Property_Area"] = le.fit_transform(test_data["Property_Area"])
```

```python
test_data.head()
```

```python
test = test_data.drop('Loan_ID',axis =1)
```

```python
cols = ["ApplicantIncome","CoapplicantIncome","LoanAmount","Loan_Amount_Term"]
```

```python
for col in cols:
  test[col] = std.fit_transform(test[col].values.reshape(-1,1))
```

```python
test.head()
```

```python
y = train["Loan_Status"]
x = train.drop(["Loan_Status"],axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=42)
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
model = lr.fit(x_train,y_train)
```

```python
pred_lr = model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
```

```python
confusion_matrix(y_test,pred_lr)
```

```python
pred_lr
```

```python
print('Accuracy =',accuracy_score(y_test,pred_lr))
```

### KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
metric_k = []
neighbors = np.arange(3,15)
```

```python
for k in neighbors:
    classifier = KNeighborsClassifier(n_neighbors = k, metric = 'euclidean')
    classifier.fit(x_train,y_train)
    y_pred = classifier.predict(x_test)
    acc = accuracy_score(y_test,y_pred)
    metric_k.append(acc)
```

```python
plt.plot(neighbors, metric_k, 'o-')
plt.xlabel('K')
plt.ylabel('Accuracy')
plt.grid()
```

```python
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'euclidean')
classifier.fit(x_train,y_train)
y_pred = classifier.predict(x_test)
acc = accuracy_score(y_test,y_pred)
metric_k.append(acc)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print('Accuracy =',accuracy_score(y_test,y_pred))
```
